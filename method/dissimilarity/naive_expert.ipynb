```python
import pandas as pd
all_profile = pd.read_csv('processed_data.csv', index_col=0)
# load naive data
naive_data = pd.read_csv('naive_set.csv', index_col=0)
# load expert data
expert_data = pd.read_csv('expert_set.csv', index_col=0)
```

```python
naive_profiles = naive_data.merge(all_profile, left_index=True, right_index=True, how='left')
expert_profiles = expert_data.merge(all_profile, left_index=True, right_index=True, how='left')
```

```python
naive_profiles.to_csv('naive_profiles.csv')
expert_profiles.to_csv('expert_profiles.csv')
```

```python
import matplotlib.pyplot as plt
top_naive = {}
top_expert = {}

col_interest = ['age_bin','gender','location','ethnicity','occupation','marital_status','children','religion','sexual_orientation']

for col in col_interest:
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
    naive_profiles[col].value_counts(normalize=True).head(5).plot(kind='bar', ax=axes[0], title=f"Naive - {col}")
    top_naive[col] = naive_profiles[col].value_counts(normalize=True).head(5)
    expert_profiles[col].value_counts(normalize=True).head(5).plot(kind='bar', ax=axes[1], title=f"Expert - {col}")
    top_expert[col] = expert_profiles[col].value_counts(normalize=True).head(5)
    plt.show()
```
